## Splitting the Dataset into independent and dependent

```python
# independent columns as my x

x = df.drop('Class', axis = 1)
```

```python
x.head()
```

```python
# Dependent columns as Y

y = df['Class']
```

```python
y
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y ,test_size = 0.2, random_state = 42)
```

```python
# checking the shape of our data
print( x_train.shape,y_train.shape,y_test.shape)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_resampled,y_resampled = smote.fit_resample(x_train,y_train)
```

```python
y_resampled.value_counts()
```

```python
# Import libraries for model training and evaluation

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from xgboost import XGBClassifier
```

```python
# Define Numeric Columns
numeric_features = x.columns.tolist()

# Preprocessing Step
preprocessor = ColumnTransformer(transformers=[('num', StandardScaler(), numeric_features)])
```

```python
# Build Pipeline with RandomForest
random_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])
```

```python
# fit the pipeline
random_pipeline.fit(x_resampled,y_resampled )
```

```python
# Predict
random_pred = random_pipeline.predict(x_test)
random_proba = random_pipeline.predict_proba(x_test)[:, 1]
```

```python
# Evaluate
print("Confusion Matrix:\n", confusion_matrix(y_test, random_pred))
print("\nClassification Report:\n", classification_report(y_test, random_pred))
print("\nROC AUC Score:", roc_auc_score(y_test, random_proba))
```

```python
rm_y_proba = random_pipeline.predict_proba(x_test)[:, 1]
# Confusion Matrix
random_cm = confusion_matrix(y_test, random_pred)

# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, rm_y_proba)
roc_auc = roc_auc_score(y_test, rm_y_proba)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Confusion Matrix Plot
sns.heatmap(random_cm, annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title('Confusion Matrix')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

# ROC AUC Curve Plot
axes[1].plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
axes[1].plot([0, 1], [0, 1], linestyle='--', color='gray')
axes[1].set_title('ROC AUC Curve')
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].legend()

plt.tight_layout()
plt.show()
```
